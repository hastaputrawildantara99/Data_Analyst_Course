## **Accessing the data**
### **Variable (columns)**
#### **Notes**
* Dalam DataFrame kolom disebut dengann variable
* Untuk melihat isi variable, gunakan `df['nama_kolom']` atau `df.nama_kolom`



```python
df[['job_title_short', 'job_location']]  # untuk melihat isi beberapa kolom
```

### **Observation (rows)**
**Notes**
* Dalam DataFrame, baris/rows disebut dengan Observations
* Untuk melihat isi dalam suatu baris, gunakan `df.iloc[angka_index]`

```python
df.iloc[0]
```

```python
df.head() # Cek isi 5 baris awal
```

**Untuk melihat isi beberapa baris di tengah, gunakan `df.iloc[n, m]`. Layar akan menampilkan list data dari observation index ke n hingga m-1**

```python
df.iloc[2:5] # Untuk melihat data baris index ke 2 hingga 4
```

## **Filtering Rows**
syntax : `df[df['nama_kolom'] > value]` berdasarkan operasi perbandingan (<,>,dll)


```python
df[df.salary_year_avg > 100000]
```

```python
df[df.job_title_short == 'Data Scientist']
```

##**Find not NA values**
