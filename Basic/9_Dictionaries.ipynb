#Dictionaries

Here's a quick comparison between these 4 container data types:

| Feature          | List                                  | Dictionary                           | Set                                | Tuple                             |
|------------------|---------------------------------------|--------------------------------------|------------------------------------|-----------------------------------|
| Syntax           | `[item1, item2, ...]`                 | `{'key1': value1, 'key2': value2}`   | `{item1, item2, ...}`              | `(item1, item2, ...)` or `item,`  |
| Order            | Ordered                               | Unordered                            | Unordered                          | Ordered                           |
| Indexing         | Yes (by index)                        | Yes (by key)                         | No                                 | Yes (by index)                    |
| Duplicate Values | Allowed                               | Values can be duplicated, keys cannot| Not allowed                        | Allowed                           |
| Mutability       | Mutable                               | Mutable                              | Mutable                            | Immutable                         |
| Usage            | For a collection of ordered items     | For key-value pairs                  | For unique items                   | For fixed data                    |

##Create a Dictionary

```python
job_type_skills = {
  'database': 'postgresql',
  'language': 'python',
  'library': 'pandas'
}
```

```python
job_type_skills
```

```python
# help(dict)
```

##Get Items
Untuk mendapatkan item pada dictionary terdapat 2 cara

###Keys
Gunakan nama Key untuk mendapatkan value

```python
job_type_skills['database']
```

###Get() function

```python
job_type_skills.get('database')
```

Gunakan function keys() untuk melihat semua keys yang ada di dictionary

```python
job_type_skills.keys()
```

```python
job_type_skills.get('analytics', 'not found') # mengembalikan 'not found' jika key yang dipanggil tidak ada, alih-alih tidak ada keterangan
```

###Values() function


```python
job_type_skills.values() # untuk melihat semua value pada dictionary
```

###Items() function

```python
job_type_skills.items() # mengembalikan pasangan key-value pada dictionary
```

###Pop() function

```python
job_type_skills.pop('library') # menghapus salah satu key dan mengembalikan value
```

```python
job_type_skills
```

## Add items
#### 1. Direct Assignment

```python
job_type_skills['framework'] = 'flask'
job_type_skills
```

#### 2. Update() function
bisa menambah beberapa item sekaligus

```python
job_type_skills.update({'cloude': 'google cloud', 'version_control': 'git'})
job_type_skills
```

## Object apapun bisa disimpan dalam dictionary

```python
job_type_skills.update({'language': ['python', 'sql']})
job_type_skills
```
