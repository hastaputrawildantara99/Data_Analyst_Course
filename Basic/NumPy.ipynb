```python
import numpy as np
```

```python
years_of_experience = np.array([1,2,3,4,5])
years_of_experience
```

##Basic operation

###1. Addition

```python
years_of_experience_plus_one = years_of_experience + 1
years_of_experience_plus_one
```

###2. Subtraction

```python
years_of_experience_minus_one = years_of_experience - 1
years_of_experience_minus_one
```

###3. Division

```python
years_of_experience_half = years_of_experience / 2
years_of_experience_half
```

###4. Multiplication

```python
years_of_experience_double = years_of_experience * 2
years_of_experience_double
```

###5. Slicing

```python
second_and_third_jobs_experience = years_of_experience[1:3]
second_and_third_jobs_experience
```

###6. Boolean Indexing

```python
jobs_with_more_than_one_year_exp = years_of_experience[years_of_experience >= 2]
jobs_with_more_than_one_year_exp
```

##Math Operation
Aggregate funct:
- `sum` = sum
- `prod` = product
- `cumsum` = cumulative sum
- `cumprod` = cumulative product  

Mathematical operations:
- `sqrt`
- `exp`
- `log`
- `sin`
- `cos`

```python
import random

salary = [random.randint(100000, 150000) for num in range(10)]
salary
```

```python
salary_array = np.array(salary)
salary_array
```

```python
# SUM
total_sum_salaries = np.sum(salary_array)
total_sum_salaries
```

```python
# Product
product_salaries = np.prod(salary_array)
product_salaries
```

```python
#CUMSUM
cumulative_sum_salaries = np.cumsum(salary_array)
cumulative_sum_salaries
```

```python
#CUMPROD
cumulative_prod_salaries = np.cumprod(salary_array)
cumulative_prod_salaries
```

##Statistic Operation
- `mean`
- `median`
- `var` : variance
- `std` : standard deviation
- `min`
- `max`

```python
# mean
average_salary = np.mean(salary_array)
average_salary
```

```python
# median
median_salary = np.median(salary_array)
median_salary
```

```python
# var
salary_variance = np.var(salary_array)
salary_variance
```

```python
# std
salary_std_dev = np.std(salary_array)
salary_std_dev
```

```python
# min
min_salary = np.min(salary_array)
min_salary
```

```python
# max
max_salary = np.max(salary_array)
max_salary
```

##Missing Value / NaN

```python
salary_with_nan = np.array([100000, np.nan, 200000, 3000000, np.nan])
salary_with_nan
```

##Replace Value

```python
salary_with_nan[salary_with_nan < 130000] = np.nan
salary_with_nan
```

##Where
`np.where(condition, if_true_condition, if_false_condition)`

```python
salary_array = np.where(salary_array < 120000, 120000, salary_array)
salary_array
```

##Random Sampling
Generate a random numbers/data or samples.
- `np.random.normal`
> specify argumen:
> - `loc`: this is the mean ( Î¼ ) of the normal (Gaussian) distribution
> - `scale`: this is the std dev of the normal (Gaussian) distribution, representing the dispersion from the mean.
> - `size`: This defines the number of random samples to draw, which is set to match the number of job posting.
> - Syntax: `np.random.normal(loc=0.0, scale=1.0, size=None)`
- `np.random.rand`
- `np.random.randn`
- `np.random.randint`
- `np.random.random`
- `np.random.uniform`
- `np.random.binomial`
- `np.random.poisson`

```python
# Generate numbers based on normal distribution
noise = np.random.normal(0, 5000, 10)
salary_array_with_noise = salary_array + noise
salary_array_with_noise
```
