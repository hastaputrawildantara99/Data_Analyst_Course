## Create Class

```python
class Lukelist :
  def __init__(self):
    """Initialize an empty list."""
    self._items = []

  def add(self, item):
    """add an item to the end of the list."""
    self._items.append(item)

  def getitem(self, index):
    """retrieve an item by index."""
    return self._items[index]

  def setitem(self, index, item):
    """Set an item at a spesific index."""
    self._items[index] = item

  def __repr__(self):
    """return a string representation of the list."""
    return str(self._items)

  def __len__(self):
    """return the length of the list."""
    return len(self._items)
```

##Create Instance

```python
my_list = Lukelist()
my_list
```

##Manipulate Instance

```python
my_list.add('Data Science')
my_list.add('Machine Learning')

print(my_list)
print(len(my_list))
```

## Practical Demonstration

```python
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=.1, symbol = '$'):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = self.base_salary * (1 + self.bonus_rate)
    self.bonus = self.total_salary - self.base_salary

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary = BaseSalary(1000)

salary.show_salary()
```

##`__str__()` function

```python
class DataScienceJobsList:
  def __init__(self, jobs):
    self.jobs = jobs

  def __str__(self):
    """Return a string representation of the lisg."""
    jobs_str = 'Data Science Job:\n'
    for job in self.jobs:
      jobs_str += f"- {job['job_title']}: {job['job_skills']}\n"
    return jobs_str
```

```python
jobs_list = DataScienceJobsList(data_science_jobs)

print(jobs_list)
```

##Object Methods

```python
class DataScienceJobsList:
  def __init__(self, jobs):
    self.jobs = jobs

  def __str__(self):
    jobs_str = 'Data Science Jobs:\n'
    for job in self.jobs:
      jobs_str += f"- {job['job_title']}: {', '.join(job['job_skills'])}\n"
    return jobs_str

  def split_skills(self):
    for job in self.jobs:
      job['job_skills'] = job['job_skills'].split(', ')
```

```python
jobs_list = DataScienceJobsList(data_science_jobs)
jobs_list.split_skills()

print(jobs_list)
```

```python
class DataScienceJobsList:
  def __init__(self, jobs):
    self.jobs = jobs

  def __str__(self):
    jobs_str = 'Data Science Jobs:\n'
    for job in self.jobs:
      if isinstance(job['job_skills'], list):
        skills_str = ', '.join(job['job_skills'])
      else:
        skills_str = job['job_skills']
      jobs_str += f"- {job['job_title']}: {skills_str}\n"
    return jobs_str

  def split_skills(self):
    for job in self.jobs:
      if isinstance(job['job_skills'], str):
        job['job_skills'] = job['job_skills'].split(', ')

  def display_jobs(self):
    for job in self.jobs:
      print(f'Job Title: {job['job_title']}')
      print("Required Skills: ")
      for skill in job['job_skills']:
        print(f'- {skill}')
      print()
```

```python
jobs_list = DataScienceJobsList(data_science_jobs)
jobs_list.split_skills()
jobs_list.display_jobs()
```
